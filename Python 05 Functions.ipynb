```python
def countdown(n):
      if n <= 0:
       print("Done!")
      else:
        print(n)
        countdown(n - 1)

countdown(5)
```

```python
def fibonacci(n):
      if n <= 1:
       return n
      else:
       return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(7))
```

```python
def my_generator():
      yield 1
      yield 2
      yield 3

for value in my_generator():
  print(value)
```

```python
def count_up_to(n):
      count = 1
      while count <= n:
        yield count
        count += 1

for num in count_up_to(5):
  print(num)
```

```python
def simple_gen():
      yield "Emil"
      yield "Tobias"
      yield "Linus"

gen = simple_gen()
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def fibonacci():
      a, b = 0, 1
      while True:
       yield a
       a, b = b, a + b

gen = fibonacci()
for _ in range(10): 
  print(next(gen))
```

```python
def my_gen():
    try:
      yield 1
      yield 2
    finally:
     print("Generator closed")

gen = my_gen()
print(next(gen))
gen.close()
```
